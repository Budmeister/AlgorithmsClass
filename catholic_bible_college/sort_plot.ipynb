```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
frame = pd.read_csv("results.csv", index_col=0)
sizes = frame["Size"]
ks = frame["k"]

results = [(column, frame[column]) for column in frame.columns if column not in ("Size", "k")]
[column for (column, _) in results]
```

```python
for name, data in results:
    plt.plot(sizes, data)
    plt.xlabel("Size")
    plt.ylabel("Time (s)")
    plt.title(f"Times for {name} versus size")
    plt.show()
```
